## Top University Analysis

```python
top_uni=df.sort_values(by='world_rank',ascending=True).drop_duplicates('institution').head(10)
top_uni[['institution','world_rank','year']]
```

```python
top_ranked = df[df['world_rank']<=100]
T=top_ranked.groupby('country')['score'].sum().sort_values(ascending=False)
T = T.reset_index()
T
```

```python
plt.figure(figsize=(20,10))
sb.barplot(x=T.country,y=T.score,palette='ocean')
plt.title("Distribution Of Top Universities Across Countries")
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Score")
plt.show()
```

## Country Performance

```python
plt.figure(figsize=(20,10))
sb.countplot(x=top_ranked.country,palette='ocean')
plt.title("Country Vs No of Top Universities")
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("No of universities")
plt.show()
```

## Factor Analysis

```python
r=df.corr(numeric_only=True)
plt.figure(figsize=(20,10))
sb.heatmap(r,annot=True,cmap='ocean')
plt.show()
```

```python
df_5=df[["publications","citations","patents","broad_impact","score"]]
sb.pairplot(df_5)
plt.show()
```

```python
c=df[["quality_of_education","alumni_employment","quality_of_faculty","world_rank"]].corr()
sb.heatmap(c,annot=True,cmap='ocean')
plt.show()
```

```python
p=pd.pivot_table(top_20,index='year',columns='institution',values='world_rank',margins=True)
plt.figure(figsize=(30,9))
sb.heatmap(p,annot=True)
plt.show()
```
